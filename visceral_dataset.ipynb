```python
#find unique value

# Find unique values in the 'number' column
unique_numbers = df['number'].unique()
print("Unique numbers:", unique_numbers)

# Find unique values in the 'age (years)' column
unique_ages = df['age (years)'].unique()
print("Unique ages:", unique_ages)

# Find unique values in the 'ethnicity' column
unique_ethnicities = df['ethnicity'].unique()
print("Unique ethnicities:", unique_ethnicities)

# Find unique values in the 'diabetes mellitus' column
unique_diabetes = df['diabetes mellitus'].unique()
print("Unique diabetes mellitus values:", unique_diabetes)
```
